## 2. Extract and display the middle section of the image (150 pixels).

```python
# Get the original image dimensions
width, height = image_grayscale.size

# Calculate the center of the image
center_x = width // 2
center_y = height // 2

# Define the crop size (150x150) and compute the bounding box
half_width = 150 // 2
left = center_x - half_width
top = center_y - half_width
right = center_x + half_width
bottom = center_y + half_width

# Crop the image using the computed coordinates
cropped_image = image_grayscale.crop((left, top, right, bottom))

# Display the cropped image
plt.imshow(cropped_image, cmap='gray')  # Show in grayscale
plt.axis('off')  # Hide axes for a cleaner look
plt.title("Cropped Image (150x150)")  # Add a title for clarity
plt.show()
```

## 3. Apply a simple threshold to the image (e.g., set all pixel values below 100 to 0).

```python
# Convert the grayscale image to a NumPy array for processing
data = np.array(image_grayscale)

# Apply a binary threshold:
# Pixels with intensity less than 100 are set to 0 (black)
# Pixels with intensity 100 or greater are set to 255 (white)
binary_data = np.where(data < 100, 0, 255)

# Convert back to an image
binary_image = Image.fromarray(binary_data.astype(np.uint8))

# Display the binary image
plt.imshow(binary_image, cmap='gray') # Show the image in grayscale mode
plt.axis('off')  # Hide axes
plt.show()
```

## 2. Eigen Decomposition and Identifying Principal Components

- **Compute Eigenvalues and Eigenvectors**.
- **Sort the eigenvalues** in descending order and choose the top k eigenvectors corresponding to the highest eigenvalues.
- **Identify the Principal Components** with the help of a cumulative sum plot.


```python
# Convert the image to a NumPy array with floating-point values
data = np.array(image, dtype=np.float32)

# Standardize the data by subtracting the mean
standardized_data = data - np.mean(data)

# Compute the covariance matrix of the standardized data
cov_matrix = np.cov(standardized_data, rowvar=False)
eigenvalues, _ = np.linalg.eigh(cov_matrix) # Perform eigen decomposition (get eigenvalues)

# Sort eigenvalues in descending order and compute cumulative variance explained
cumulative_sum = np.cumsum(eigenvalues[::-1]) / np.sum(eigenvalues)

# Plot the cumulative explained variance
plt.plot(cumulative_sum, marker='o', linestyle='--')  # Line plot with markers
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance')
plt.title('Cumulative Explained Variance')
plt.grid()
plt.show()

# Display the original image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()  # Show the image
```

## 3. Reconstruction and Experiment:

- **Reconstruction**: Transform the original data by multiplying it with the selected eigenvectors (PCs) to obtain a lower-dimensional representation.

- **Experiments**: Pick four different combinations of principal components with various explained variance values and compare the results.

- **Display the Results and Evaluate**.


```python
# Convert image to a NumPy array with floating point values
data = np.array(image, dtype=np.float32)

# Standardize data
mean = np.mean(data)
data -= mean

# Compute the covariance matrix of the standardized data
cov_matrix = np.cov(data, rowvar=False)
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix) # Get the eigenvalues and eigenvectors of the covariance matrix
idx = np.argsort(eigenvalues)[::-1]
eigenvectors = eigenvectors[:, idx]

# Define different numbers of principal components to use for reconstruction
components = [10, 20, 50, 100, 150]
fig, axes = plt.subplots(1, 6, figsize=(18, 5)) # Create a figure with multiple subplots to display images
# Show the original image in the first subplot
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original main")
axes[0].axis("off")

# Reconstruct the image using different numbers of principal components
for i, k in enumerate(components):
    reconstructed = (data @ eigenvectors[:, :k] @ eigenvectors[:, :k].T) + mean # Project data onto top k eigenvectors and reconstruct the image
    # Display the reconstructed image in the corresponding subplot
    axes[i+1].imshow(reconstructed, cmap='gray')
    axes[i+1].set_title(f" {k}:Components")
    axes[i+1].axis("off")  # Hide axes for better visualization
```
