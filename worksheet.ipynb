## Task 1. Classify Temperatures

```python
# List of temperatures recorded at Tribhuwan International Airport
temperatures = [
    8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
    16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3,
    13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7,
    7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3,
    16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5
]

# Create empty lists for classification
cold = []
mild = []
comfortable = []

# Classify temperatures
for temp in temperatures:
    if temp < 10:
        cold.append(temp)
    elif 10 <= temp < 15:
        mild.append(temp)
    elif 15 <= temp < 20:
        comfortable.append(temp)

# Display classified temperatures
print("Cold temperatures:", cold)
print("Mild temperatures:", mild)
print("Comfortable temperatures:", comfortable)
```

## Task 2: Based on Data

```python
# Count the number of occurrences in each category
mild_count = len(mild)
comfortable_count = len(comfortable)
cold_count = len(cold)

# Print the results
print("Number of mild days:", mild_count)
print("Number of comfortable days:", comfortable_count)
print("Number of cold days:", cold_count)
```

## Task 3. Convert Temperatures from Celsius to Fahrenheit

```python
# List of temperatures measured at Tribhuwan International Airport
temperatures = [8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
                16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3,
                13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7,
                7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3,
                16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5]

# Create an empty list for Fahrenheit temperatures
temperatures_fahrenheit = []

# Convert Celsius to Fahrenheit
for temp in temperatures:
    fahrenheit = (temp * 9/5) + 32
    temperatures_fahrenheit.append(fahrenheit)

# Print the converted Fahrenheit values
print("Temperatures in Fahrenheit:", temperatures_fahrenheit)
```

## Task 4. Analyze Temperature Patterns by Time of Day

```python
import matplotlib.pyplot as plt

# List of temperatures measured at Tribhuwan International Airport
temperatures = [8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
                16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3,
                13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7,
                7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3,
                16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5]

# Task 4: Analyze Temperature Patterns by Time of Day
night_temps = []
evening_temps = []
day_temps = []

# Assign temperatures to the appropriate time-of-day lists
for i in range(len(temperatures)):
    if i % 3 == 0:  # Night temperatures (0, 3, 6, ...)
        night_temps.append(temperatures[i])
    elif i % 3 == 1:  # Evening temperatures (1, 4, 7, ...)
        evening_temps.append(temperatures[i])
    elif i % 3 == 2:  # Day temperatures (2, 5, 8, ...)
        day_temps.append(temperatures[i])

# Calculate the average daytime temperature
average_day_temp = sum(day_temps) / len(day_temps)

# Print the average daytime temperature
print("Average daytime temperature:", average_day_temp)

# Optional: Plot "day vs. temperature"
plt.figure(figsize=(10, 5))
plt.plot(range(len(day_temps)), day_temps, marker='o', linestyle='-', color='b')
plt.title('Daytime Temperatures at Tribhuwan International Airport')
plt.xlabel('Day Index')
plt.ylabel('Temperature (°C)')
plt.xticks(range(len(day_temps)))  # Set x-ticks to show each day
plt.grid()
plt.show()
```

Notes

- **X-axis (Day Index)**: Represents the sequence of days.
- **Y-axis (Temperature in °C)**: Represents recorded temperatures in degrees Celsius.
- **Data Representation**:
  - A blue line connects the temperature points across days.
  - Blue circular markers indicate individual temperature readings.
- **Grid Lines**: Present for better readability.
- **Average Temperature**: Displayed at the top of the plot, which is calculated as 13.40625°C.


## 8.1 Recursion

1.Sum of Nested Lists (Recursion)

```python
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total

# Test
print(sum_nested_list([1, [2, [3, 4], 5], 6, [7, 8]]))  # Output: 36
```

2. Fibonacci Sequence (Recursion)

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Test
print(fibonacci(5))  # Output: 5
```

3. Reverse a String (Recursion)

```python
def reverse_string(s):
    if len(s) == 0:
        return s
    else:
        return s[-1] + reverse_string(s[:-1])

# Test
print(reverse_string("hello"))  # Output: "olleh"
```

4. Tower of Hanoi (Recursion)

```python
def tower_of_hanoi(n, source, target, auxiliary):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    tower_of_hanoi(n - 1, source, auxiliary, target)
    print(f"Move disk {n} from {source} to {target}")
    tower_of_hanoi(n - 1, auxiliary, target, source)

# Test
tower_of_hanoi(3, 'A', 'C', 'B')
```

5. Binary Search (Divide and Conquer)

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Test
print(binary_search([1, 2, 3, 4, 5], 3))  # Output: 2
```

6. Merge Sort (Divide and Conquer)

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

# Test
arr = [38, 27, 43, 3, 9, 82, 10]
merge_sort(arr)
print(arr)  # Output: [3, 9, 10, 27, 38, 43, 82]
```

7. Permutations of a String (Recursion)


```python
def permutations(s):
    if len(s) == 1:
        return [s]
    perm_list = []
    for i, char in enumerate(s):
        for perm in permutations(s[:i] + s[i+1:]):
            perm_list.append(char + perm)
    return perm_list

# Test
print(permutations("abc"))  # Output: ['abc', 'acb', 'bac', 'bca', 'cab', 'cba']
```

8. Depth-First Search (DFS) in a Graph (Recursion)

```python
def dfs(graph, node, visited=None):
    if visited is None:
        visited = set()
    visited.add(node)
    print(node)
    for neighbor in graph[node]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

# Test
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}
dfs(graph, 'A')  # Output: A B D E F C
```
